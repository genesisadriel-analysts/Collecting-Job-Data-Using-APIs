```python
import requests

api_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json"

def get_number_of_jobs_T(Python):
    response = requests.get(api_url)
    data = response.json()

    number_of_jobs = 0

    for job in data:
        if "Python" in job["Key Skills"]:
            number_of_jobs += 1

    return number_of_jobs
```

Create a python list of all technologies for which you need to find the number of jobs postings.

```python

# List of technologies
technologies = [
    "C", "C#", "C++", "Java", "JavaScript", 
    "Python", "Scala", "Oracle", "SQL Server", 
    "mysql", "PostgreSQL", "MongoDB"
]

# Function to count jobs for a given technology
def get_number_of_jobs_for_tech(tech):
    response = requests.get(api_url)
    data = response.json()
    count = 0
    for job in data:
        # Check if the technology exists in "Key Skills" (case-insensitive)
        if tech.lower() in job["Key Skills"].lower():
            count += 1
    return count

# Collect results
results = []
for tech in technologies:
    count = get_number_of_jobs_for_tech(tech)
    results.append({"Technology": tech, "Number of Jobs": count})

# Create a DataFrame
df = pd.DataFrame(results)

# Write to Excel
df.to_excel("jobs_by_technology.xlsx", index=False)

print("Results saved to jobs_by_technology.xlsx")
```

Find the number of jobs postings for each of the technology in the above list.
Write the technology name and the number of jobs postings into the excel spreadsheet.

```python
# List of technologies
technologies = [
    "C", "C#", "C++", "Java", "JavaScript",
    "Python", "Scala", "Oracle", "SQL Server",
    "mysql", "PostgreSQL", "MongoDB"
]

# Call API
response = requests.get(api_url)
data = response.json()  # data is likely a dictionary

# If the jobs are under a key like 'jobs', get the list:
# jobs_list = data['jobs']   # Uncomment if your JSON has 'jobs' key
# If data itself is already a list, just use data directly:
jobs_list = data

# Create Excel workbook
wb = Workbook()
ws = wb.active
ws.title = "Job Postings"

# Add header
ws.append(["Technology", "Number of Jobs"])

# Count jobs for each technology
for technology in technologies:
    number_of_jobs = 0
    for job in jobs_list:
        # Case-insensitive search
        if technology.lower() in job["Key Skills"].lower():
            number_of_jobs += 1
    ws.append([technology, number_of_jobs])

# Save workbook
wb.save("jobs_by_technology.xlsx")
print("Saved jobs_by_technology.xlsx")
```
